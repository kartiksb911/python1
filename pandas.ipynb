# grouping of data

```python
data = {
  'co2': [95, 90, 99, 104, 105, 94, 99, 104],
  'model': ['Citigo', 'Fabia', 'Fiesta', 'Rapid', 'Focus', 'Mondeo', 'Octavia', 'B-Max'],
  'car': ['Skoda', 'Skoda', 'Ford', 'Skoda', 'Ford', 'Ford', 'Skoda', 'Ford']
}

df3 = pd.DataFrame(data)
```

```python
df3
```

```python
df4=df3.groupby("car")
```

```python
df4
```

```python
for x,y in df4:
    print(x)
    print(y)
    print()
```

```python
df4.get_group("Ford")
```

```python
df4.max()
```

```python
df4.size()
```

```python
df4.first()
```
